## Creating a Function to Read All the Data

```python
import pandas as pd
def read_table(data1):
    table1=pd.DataFrame()
    # Dataframe of Boolean, True where the Data is not None/Null
    list1=data1.notnull()
    Documents=[]
    Descriptions=[]
    rev=[]
    Status=[]
    list2=[]
    list3=[]
    print(len(data))
    #Start the loop the using the second row of the Tabula Extracted DataFrame
    for i in range(1,len(data1)):
        doc=''
        des=''
        # if the First item in the row is not null
        if list1.iloc[i,0]:
            startnum=i
            endnum=i
            
            # finding the end row number where the first entry ends:
            if endnum+1<len(list1):
                #Run the loop until the next non empty cell is reached:
                while ~list1.iloc[(endnum+1)%len(list1),0]:
                    if endnum+1<len(list1):
                        endnum+=1
                    else:
                        break

            #concatenate the Text for the determined rows and store them in respective columns as lists.       
            for j in range(startnum,endnum+1):
                list2.append(j)
                doc+=str(data1.iloc[j,1])
                #if the Description is the second column then follow this strategy 
                if data1.iloc[0,2]=='D E S C R I P T I O N':
                    if list1.iloc[j,2]:
                        des+=data1.iloc[j,2]
                else:
                    if list1.iloc[j,2]:
                        des+=data1.iloc[j,2]
                        if list1.iloc[j,3]:
                            des+=data1.iloc[j,3]
        if doc!='':
            Documents.append(doc)
        if des!='':
            Descriptions.append(des)
    table1['Documents']=Documents
    table1['Descriptions']=Descriptions
    return table1
        
```
